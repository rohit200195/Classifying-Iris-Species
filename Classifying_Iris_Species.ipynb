```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(iris_dataset['data'],iris_dataset['target'],random_state = 0)
```

```python
print("x_train shape : {}".format(X_train.shape))
print("y_train shape : {}".format(y_train.shape))
```

```python
print("x_test shape : {}".format(X_test.shape))
print("y_test shape : {}".format(y_test.shape))
```

```python
import pandas as pd
iris_dataframe = pd.DataFrame(X_train,columns = iris_dataset.feature_names)
iris_dataframe.head()
```

```python
iris = sns.load_dataset("iris")
g = sns.pairplot(iris, hue="species",height = 3.5,plot_kws=dict(s=100, edgecolor="b", linewidth=1))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train,y_train)
```

```python
X_new = np.array([[5, 2.9 ,1 ,0.2]])
print('shape is : {}'.format(X_new.shape))
```

```python
prediction = knn.predict(X_new)
print('prediction : {}'.format(prediction))
print('prediction name is  : {}'.format(iris_dataset['target_names'][prediction]))
```

```python
y_pred = knn.predict(X_test)
print('Test set predictions : \n {}'.format(y_pred))
```

```python
y_test
```

```python
print("Test score : {:.2f}".format(np.mean(y_pred == y_test)))
```

```python
#using score method 
print("Test score is : {:.2f}".format(knn.score(X_test,y_test)))
```
